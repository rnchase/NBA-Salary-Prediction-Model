```python
import pandas as pd
import numpy as np
import matplotlib as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.compose import make_column_transformer
from sklearn.compose import make_column_selector
import matplotlib.pyplot as plt 
from sklearn.metrics import mean_squared_error
```

```python
# Bringing in data sets
PlayerStats = pd.read_csv('Player Stats 22-23.csv', index_col=0, encoding = 'latin-1')
Salaries = pd.read_csv('Salaries.csv', index_col = 0, encoding = 'latin-1')



# Merging player statistics and salary 

PlayerStats = PlayerStats.merge(Salaries, how = 'outer', on = 'Player')


pd.set_option('display.max_rows', None)
#print(PlayerStats)

Salaries['Player'] = Salaries['Player'].astype("string")
PlayerStats['eFG%'] = PlayerStats['eFG%'].astype(float)

```

```python

# replace NaN salary values with 0
PlayerStats['2023-24'] = PlayerStats['2023-24'].fillna(0)
PlayerStats['2024-25'] = PlayerStats['2024-25'].fillna(0)
PlayerStats['2025-26'] = PlayerStats['2025-26'].fillna(0)
PlayerStats['2026-27'] = PlayerStats['2026-27'].fillna(0)
PlayerStats['2027-28'] = PlayerStats['2027-28'].fillna(0)
PlayerStats['2028-29'] = PlayerStats['2028-29'].fillna(0)

# Removing Players with No statistics
PlayerStats.dropna(subset = ['Pos'], inplace=True)


# Changing NaN statistic with 0
PlayerStats.fillna(0,inplace=True)
PlayerStats = PlayerStats[PlayerStats['2023-24'] != 0]



```

```python
# Linear Regression Testing 
X = PlayerStats[["PTS"]].to_numpy()
y = PlayerStats["2023-24"].to_numpy()
X = X.ravel()
y = y.ravel()

mymodel = np.poly1d(np.polyfit(X, y,5))
myline = np.linspace(0, 35, 100)



plt.scatter(X, y)
plt.plot(myline, mymodel(myline))
plt.xlabel('Points Per Game') #x label
plt.ylabel('Salary in Tens of Thousands') #y label
```
